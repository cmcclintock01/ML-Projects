## Data Set 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ds_1 = pd.read_csv("ds10.csv")
ds_1.describe()
```

```python
#Here is my plt to show that data
plt.scatter(ds_1['x'], ds_1['y'], color='red')
plt.show()
#import KMeans, create the classifier, and fit the data
from sklearn.cluster import KMeans
X = ds_1[['x', 'y']]

kmeans_classifier = KMeans(n_clusters=6)
kmeans_classifier.fit(X)
#import math and evaluate
import math
print("The distortion is ", kmeans_classifier.inertia_)

RMS_average = math.sqrt(kmeans_classifier.inertia_ / len(X))
print("Averaging and take the square root gives ", RMS_average)

cluster_x = [x for [x,y] in kmeans_classifier.cluster_centers_]
cluster_y = [y for [x,y] in kmeans_classifier.cluster_centers_]
plt.scatter(ds_1['x'], ds_1['y'], c=kmeans_classifier.labels_)
plt.scatter(cluster_x, cluster_y, marker="+", color="red")
plt.show()
```

I believe that there was 6 clusters created in this data set because the distortion score stops decreasing dramatically after 6 clusters. From 5 clusters to 6 clusters there was a change of over 200 in the distortion score while from 6 to 7 clusters there was only a change of 50.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ds_3 = pd.read_csv("ds12.csv")
ds_3.describe()
```

```python

fig = plt.figure()

# HERE IS WHERE WE CREATE THE 3D PLOT
ax = fig.add_subplot(111, projection='3d')

ax.scatter(ds_3['x'], ds_3['y'], ds_3['z'])

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
from sklearn.cluster import KMeans
X = ds_3[['x', 'y', 'z']]

kmeans3_classifier = KMeans(n_clusters=8)
kmeans3_classifier.fit(X)
#import math and evaluate
import math
print("The distortion is ", kmeans3_classifier.inertia_)

RMS_average = math.sqrt(kmeans3_classifier.inertia_ / len(X))
print("Averaging and take the square root gives ", RMS_average)

cluster_x = [x for [x,y,z] in kmeans3_classifier.cluster_centers_]
cluster_y = [y for [x,y,z] in kmeans3_classifier.cluster_centers_]
cluster_z = [z for [x,y,z] in kmeans3_classifier.cluster_centers_]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(ds_3['x'], ds_3['y'], ds_3['z'], c=kmeans3_classifier.labels_)
ax.scatter(cluster_x, cluster_y, cluster_z, marker="+", color="red")
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()
```

## Data Set 5

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ds_5 = pd.read_csv("ds14.csv")
ds_5.describe()
```

```python
X = ds_5[['x', 'y', 'z', 'w']]
#Going to try to reduce dimensions
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
segmentation_std = scaler.fit_transform(X)
pca = PCA()
pca.fit(segmentation_std)
pca.explained_variance_ratio_

plt.figure()
plt.plot(range(1,5), pca.explained_variance_ratio_.cumsum())
plt.show()


pca = PCA(n_components = 3)
pca.fit(segmentation_std)
scores_pca = pca.transform(segmentation_std)

from sklearn.cluster import KMeans

wcss = []
for i in range(1, 21):
    kmeans_pca = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans_pca.fit(scores_pca)
    wcss.append(kmeans_pca.inertia_)

plt.figure()
plt.plot(range(1, 21), wcss)
plt.show()
```

Using this graph I would sat that there are roughly 6 clusters in this data set. The graph starts to smooth out after 5 but there is still a significant drop from 5 to 6 as compared to 6 to 7.5.
